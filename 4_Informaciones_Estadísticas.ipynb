## Media

Si queremos obtener la media de alguna variable numérica, simplemente ejecutamos el método `mean()` en la Serie. Es decir, seleccionamos esa columna en particular y ejecutamos la función`mean()`.

```python
# ver la média de la columna ride_duration
df.ride_duration.mean()
```

## Mediana

La mediana es el valor que separa la mitad superior y la mitad inferior de un conjunto de datos. Por ejemplo, considera el conjunto `mis_datos =[1,1,2,4,9].`

El valor que está "en el medio" es el número 2. No debes confundirlo con la media, que es la suma de todos los valores dividida por su cantidad.

## Desviación estándar

El desvio estándar es una medida de dispersión alrededor de la media poblacional de una variable.


 $$SD = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \overline{x}) ^ 2}$$

Hablando de manera simplificada, un desvio estándar alto indica que los datos están más dispersos en relación con su media. Por otro lado, un desvio estándar bajo indica que los valores están más agrupados, más "cerca" unos de otros.

Por ejemplo, en la figura a continuación, se muestra un ejemplo de valores altos y bajos para el desvio estándar. En rojo, los datos están más cerca de su media, y en azul están más dispersos.

Observa en la figura a continuación cómo están ubicados los disparos en cada una de las situaciones. Identifica mentalmente cuáles tienen mayor o menor dispersión, es decir, desvio estándar.

Para calcular el desvio estándar de cualquier columna, simplemente ejecutas el método `std()`en la variable.

```python
# calcular la mediana
pd.Series([1,1,2,4,900]).median()
```

```python
# calcular el desvio padron para ride_calculation
df.ride_duration.std()
```

## Valores maximos y minimos

Para calcular el valor máximo, simplemente ejecutas el método `max()`, mientras que para el valor mínimo, basta con ejecutar `min()`.

```python
# valor máximo
df.ride_duration.max()
```

```python
# valor mínimo
df.ride_duration.min()
```
