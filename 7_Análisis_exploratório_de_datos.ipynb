## Resumen del DataFrame

Ejecutando el método `info()` en una variable, obtienes un resumen general de todas las variables (colunas) de su conjunto de datos.

Puedes ver la cantidad de entradas, valores no nulos y el tipo de cada variable en una variable. También se informa la cantidad de memoria que se está utilizando para cargar el *DataFrame* en la máquina.

```python
# ver el resumen del df
df.info()
```

## Identificar el tipo de cada coluna

Los tipos de variables son información importante para incluir en el Diccionario de Variables de un proyecto, ya que garantizan una buena práctica de documentación del proyecto.

```python
# identificar los tipos de variábles
df.dtypes
```

## Valores únicos

Para ver quales son todos los valroes únicos de una columna, es conveniente usar el método `unique()`.


```python
df.ocorrencia_tipo.unique()
```

Aunque el método anterior devuelve una lista de valores únicos, a menudo queremos saber con qué frecuencia se producen.


En el ejemplo de la columna`ocorrencia_tipo`, para averiguar el número de cada tipo de ocurrencia, podemos usar `value_counts()`.

```python
df.ocorrencia_tipo.value_counts()
```

Para transformar la celda anterior en un porcentaje, basta con dividir los valores por el número total de filas de nuestro conjunto de datos.

Como vimos anteriormente, el número de filas se obtiene mediante `df.shape[0]`.

```python
df.ocorrencia_tipo.value_counts() / df.shape[0]
```

## Filtrando valores

Existen algunos métodos principales para recuperar y filtrar valores en un *DataFrame*, cada uno con sus propias características.

En lugar de intentar explicarlos, lo que resulta muy confuso para quienes nunca han utilizado ninguno de ellos, voy a dar un ejemplo de cómo se utiliza cada uno.

### loc

Para utilizar esta opción, debe entender que funciona de la siguiente manera:

`loc[<filas>, <columnas>]`.

Este método puede utilizarse en dos situaciones:

* Seleccionar filas por *etiquetas*/*índice*.
* Seleccionar filas en base a un *array booleano

Pondré un ejemplo de la primera situación en la que podemos utilizar `loc`. Supongamos que necesito extraer la latitud y longitud del lugar donde ocurrió la ocurrencia número `200808073991179`:

```python
df.loc[ 200808073991179, ['ocorrencia_latitude', 'ocorrencia_longitude']]
```

Te daré un ejemplo de la segunda situación en la que podemos usar loc. Supongamos que quiero saber todos los vuelos en los que tengo una ocurrencia del tipo "FALHA DO MOTOR EM VOO":

```python
# ver sólo las 5 primeras entradas de la situación
df.loc[df["ocorrencia_tipo"] == "FALHA DO MOTOR EM VOO"]. head()
```

### iloc

El uso de `iloc` es más sencillo, porque seleccionamos filas y columnas utilizando números, en el mismo orden en que aparecen en el *DataFrame*:

`loc[<número_de_filas>, <número_de_columnas>]`.

Puedo seleccionar un *rango* de interés. Por ejemplo, para seleccionar las 3 primeras entradas y las 4 primeras columnas, lo haría de la siguiente manera:

```python
df.iloc[:3,:4]
```
